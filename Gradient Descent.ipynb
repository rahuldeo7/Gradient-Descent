```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate synthetic data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
# Plot the synthetic data
plt.scatter(X, y)
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
# Gradient descent parameters
learning_rate = 0.1
n_iterations = 1000
m = 100
```

```python
# Initial guess for theta (theta_0 and theta_1)
theta_0 = np.random.randn()
theta_1 = np.random.randn()
```

```python
# Alternative Gradient Descent Algorithm
for iteration in range(n_iterations):
    # Initialize gradients
    gradient_0 = 0
    gradient_1 = 0
    
    for i in range(m):
        # Calculate the residual for each data point
        residual = (theta_0 + theta_1 * X[i]) - y[i]
        
        # Calculate gradients for theta_0 and theta_1
        gradient_0 += (2/m) * residual
        gradient_1 += (2/m) * residual * X[i]
    
    # Update parameters
    theta_0 = theta_0 - learning_rate * gradient_0
    theta_1 = theta_1 - learning_rate * gradient_1

print(f"Estimated coefficients: intercept (theta_0) = {theta_0}, slope (theta_1) = {theta_1}")
```

```python

# Plot the regression line
plt.scatter(X, y)
plt.plot(X, theta_0 + theta_1 * X, color='red', linewidth=2)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression with Gradient Descent")
plt.show()
```
